```python
d = {}
with open("dataset/train/1.txt") as f:
    for line in f:
      x = line.split()
      if (len(x) == 2):
        (key, val) = x
      else:
        if x[0] == "Weather":
          key = "Weather_conditions"
          val = x[2]
        elif x[0] == "Time_taken":
          key = "Time_taken_min"
          val = x[2]
        elif(len(x) == 4 and x[0] == "Name:"):
          # better error handling; ignore last name
          pass
        else:
          print(f'len > 2; {x}')
      d[key] = val
```

```python
import os
train_path='dataset/train/'
list_d = []
for filename in os.listdir(train_path):
   with open(os.path.join(train_path, filename), 'r') as f: # open in readonly mode
      # do your stuff
      # print(filename)
      d = {}
      for line in f:
        x = line.split()
        if (len(x) == 2):
          (key, val) = x
        else:
          if x[0] == "Weather":
            key = "Weather_conditions"
            val = x[2]
          elif x[0] == "Time_taken":
            key = "Time_taken_min"
            val = x[2]
          elif(len(x) == 4 and x[0] == "Name:"):
            # better error handling; ignore last name
            pass
          else:
            print(f'{filename}: len > 2; {x}')
        d[key] = val
      list_d.append(d)
```
